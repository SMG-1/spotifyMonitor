```python
from http import client
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import yaml
import pandas as pd

scope = "user-library-read"


with open("C:/Users/smira/OneDrive/Documents/CodingProjects/spotiMonitor/credentials/credentials.yml", "r") as stream:
    try:
        content = yaml.safe_load(stream)
    except yaml.YAMLError as exception:
        print(exception)

client_id=content['SPOTIFY_CLIENT_ID']
client_secret=content['SPOTIFY_SECRET_ID']
redirect_uri='http://localhost:8888/callback/'

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=client_id,
    client_secret=client_secret,
    redirect_uri=redirect_uri,
    scope=scope))

sp2 = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=client_id,
    client_secret=client_secret,
    redirect_uri=redirect_uri,
    scope='user-top-read'))
```

top tracks

```python
res = sp2.current_user_top_tracks(limit=30, time_range='long_term')['items']

df = pd.DataFrame(res)
df_album = pd.json_normalize(pd.json_normalize(df['artists'])[0])
# df_album = pd.json_normalize(df['artists'])

df_album.rename(columns={'name': 'artistName'}, inplace=True)
df = pd.concat([df, df_album[['artistName']]], axis=1)
df.rename(columns={'id': 'trackId'}, inplace=True)

df[['trackId', 'name', 'popularity', 'artistName']].to_csv('../data/top_tracks.csv', index=False)
```
